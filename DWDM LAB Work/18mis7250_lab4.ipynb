```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('cars.csv')
df
```

**solution of 1st question encoding top 10 X0 values**

```python
count=df['X0'].value_counts()
count
```

```python
top_10=count.head(10)
top_10
```

```python
encoding_top10 = top_10.copy()
encoding_top10 = pd.get_dummies(encoding_top10, columns=['X0'], prefix = ['X0'])

encoding_top10.head()
```

**solution for 2nd question label encoding on sorted class**

```python
df['Area']
df['X14']
```

```python
df['Area'] = df.Area.str[0]
df['Area'].value_counts()
```

```python
df.groupby(['Area'])['X14'].mean()
```

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
df['Area'] = lb_make.fit_transform(df['X14'])

df.head()
```
